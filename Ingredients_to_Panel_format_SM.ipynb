```python
import pandas as pd
import numpy as np

#Read the downloaded report
Path_A = "/Users/swaroopbasavarajmusti/Desktop/Nutritional_Script_data.xlsx"

Nutri_Data = pd.read_excel(Path_A)

Nutri_Data = Nutri_Data.iloc[1:]

Nutri_unpivot = pd.melt(Nutri_Data, id_vars=['_id','sectionTitle.en'], value_vars=['Calories', 'Energy','Carbohydrate','of which sugars',\
                'Fat','of which saturated fat','Protein','Salt'], var_name='itemName.en', value_name='itemValue.en')


for ind, row in Nutri_unpivot.iterrows():
    if Nutri_unpivot.loc[ind,'itemName.en'] == 'Calories':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Kalorie'
        Nutri_unpivot.loc[ind,'itemIndex'] = '0'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'kcal'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'kcal'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'Energy':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Energie'
        Nutri_unpivot.loc[ind,'itemIndex'] = '1'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'kJ'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'kJ'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'Carbohydrate':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Kohlenhydrate'
        Nutri_unpivot.loc[ind,'itemIndex'] = '2'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'g'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'g'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'of which sugars':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'davon Zucker'
        Nutri_unpivot.loc[ind,'itemIndex'] = '3'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'g'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'g'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'Fat':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Fett'
        Nutri_unpivot.loc[ind,'itemIndex'] = '4'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'g'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'g'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'of which saturated fat':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Fettsäuren'
        Nutri_unpivot.loc[ind,'itemIndex'] = '5'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'g'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'g'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'Protein':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Eiweiß'
        Nutri_unpivot.loc[ind,'itemIndex'] = '6'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'g'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'g'
    elif Nutri_unpivot.loc[ind,'itemName.en'] == 'Salt':
        Nutri_unpivot.loc[ind,'itemName.de'] = 'Salz'
        Nutri_unpivot.loc[ind,'itemIndex'] = '7'
        Nutri_unpivot.loc[ind,'tableTitle.en'] = 'Nuritional Data'
        Nutri_unpivot.loc[ind,'tableTitle.de'] = 'Nährwertangaben'
        Nutri_unpivot.loc[ind,'itemUnit.en'] = 'g'
        Nutri_unpivot.loc[ind,'itemUnit.de'] = 'g'
    else:
        pass 
    

Nutri_unpivot = Nutri_unpivot.sort_values(by=['_id','itemIndex']).reset_index(drop = True)
Nutri_unpivot[['tableTitle.tr','sectionTitle.tr','itemName.tr','itemValue.tr','itemUnit.tr']] =''
Nutri_unpivot[['tableIndex','sectionIndex']] ='0'
Nutri_unpivot['sectionTitle.de'] = Nutri_unpivot['sectionTitle.en'].str.replace('Per', 'Pro')
Nutri_unpivot['itemValue.de'] = Nutri_unpivot['itemValue.en']

Nutri_unpivot = Nutri_unpivot.applymap(str)

Nutri_unpivot = Nutri_unpivot[['_id','tableTitle.de','tableTitle.en','tableTitle.tr','tableIndex','sectionTitle.de',\
                               'sectionTitle.en','sectionTitle.tr','sectionIndex','itemName.de','itemName.en',\
                               'itemName.tr','itemValue.de','itemValue.en','itemValue.tr','itemUnit.de',\
                               'itemUnit.en','itemUnit.tr','itemIndex']]

Nutri_unpivot.to_excel("/Users/swaroopbasavarajmusti/Desktop/nutrients_output.xlsx")

```
