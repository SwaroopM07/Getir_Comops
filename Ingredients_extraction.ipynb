```python
import re
import pandas as pd

def extract_numbers_to_column(df, column_name, text, new_column_name):
    pattern = re.compile(f"{text}(r'\d+(?:,\d+)?'|(\d+))")
    extracted_numbers = []
    for value in df[column_name]:
        match = pattern.search(value)
        if match:
            extracted_numbers.append(str(match.group(1)))
        else:
            extracted_numbers.append(None)
    df[new_column_name] = extracted_numbers
    return df
```

```python
Path_A = "/Users/swaroopbasavarajmusti/Desktop/test_data.xlsx"
df = pd.ExcelFile(Path_A)
Ingredients_list = pd.read_excel(df, 'DE Product Descriptions')
Ingredients_list = Ingredients_list[['_id','description_en','ingredients_en']]
Ingredients_list = Ingredients_list.astype(str)
```

```python
Ingredients = ['Kohlenhydrate','Zucker','Fett','Fettsäuren','Eiweiß','Salz']
```

```python
for i in Ingredients:
    Ingredients_df = extract_numbers_to_column(Ingredients_list, 'ingredients_en', i , i+'.UOM')
```
